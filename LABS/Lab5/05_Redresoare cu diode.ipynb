# <font color = '#0f0264'>  1. Notiuni teoretice </font>
Redresorul este un circuit electronic care asigura conversia unei tensiuni alternative, de componenta medie nula, aplicată la intrarea sa, într-o tensiune continuă pulsatorie. Din acest motiv, se spune ca redresorul este un convertor AC $\rightarrow$ DC (A*lternative* C*urrent* $\rightarrow$ D*irect* C*urrent*). 

In general, redresoarele sunt utilizate in sursele de alimentare cu energie electrica ale sistemelor electronice. Pentru sistemele electronice uzuale sunt utilizate o varianta de redresoare, numite monofazate; aceste redresoare sunt de 2 tipuri si anume:
- redresoare monoalternanţă
- redresoare bialternanţă

## <font color = '#0f0264'> 1.1. Redresorul monofazat monoalternanţă </font>
Schema electrica a redresorului monoalternanţă este prezentată în Figura 1, in care prin $R_L$ s-a notat  **sarcina** redresorului, care este rezistenta electrica de pe care se culege tensiunea de ieşire a redresorului. Rolul redresorului este de a asigura sarcinii o tensiune electrica sau un curent electric cat mai constant. Sarcina redresorului poate fi rezistenta electrica a unui rezistor sau a oricarei alte componente electronice, sau rezistenta electrica echivalenta, masurata intre 2 puncte ale unui circuit oarecare. 

Dioda $D$ reprezintă o diodă semiconductoare care are rolul de a redresa curentul electric, care trece prin rezistenta de sarcina $R_L$. Functia de redresare a curentului electric consta in fortarea unui sens unic pentru curentul electric care trece prin sarcina redresorului. 

<img width = "300" height = "200" src="501.png" imgborder = "5">
<p style="text-align: center;">Figura 1. Redresorul monoalternanta</p>

La intrarea redresorului se aplică o tensiune sinusiodală $v_I(t)$ de amplitudine mare (în general de ordinul zecilor de volţi), a cărei componentă medie este nulă (variaţia tensiunii de intrare este axată pe zero volti). 

Comportamentul redresorului monoalternanţă depinde de regiunea în care funcţionează dioda semiconductoare:
- pe durata semialternanţei pozitive a tensiunii de intrare (durata în care $v_I$>0[V]), dioda funcţionează în conducţie directă, permiţând trecerea curentului prin ea, care ajunge pe rezistenta de sarcina $R_L$ şi generează pe aceasta o tensiune electrica a carei valoare tinde la valoarea tensiunii de intrare $v_I$; diferenta dintre valoarea tensiunii de intrare $v_I$ si valoarea tensiunii de iesire $v_O$ este data de tensiunea care cade pe rezistenta electrica a diodei, pe care se pierde o parte din valoarea tensiunii de intrare $v_I$;  
- pe durata semialternanţei negative a tensiunii de intrare (durata în care $v_I$<0[V]), dioda funcţionează în conducţie inversă şi blochează trecerea curentului prin ea, astfel încât, prin rezistenta de sarcina $R_L$ nu trece curent electric, iar tensiunea pe acesta este nulă. 

<img width = "450" height = "350" src="502.png" imgborder = "5">
<p style="text-align: center;">Figura 2. Formele de unda ale tensiunilor electrice ale redresorului monoalternanta. </p>

Forma de undă a tensiunii electrice de iesire, obtinute prin redresare, este indicata in Figura 2. Tensiunea de iesire $v_O$ a redresorului contine atat o componenta continua (DC), cat si o componenta alternativa (AC). Dintre cele 2 componente, tinand cont de scopul redresorului, componenta alternativa este de nedorit. Performanta unui redresor este masurata in functie marimea componentei alternative, raportata la componenta continua, iar parametrul care furnizeaza aceasta marime este **factorul de ondulatie (*ripple factor*)**, notat cu $\gamma$. **Un redresor este cu atat mai performant cu cat factorul de ondulatie $\gamma$ este mai mic.**

Pentru redresorul monoalternanta, factorul de ondulatie este egal cu:

$$\gamma_{monoalternanta} \approx 1.21$$ 

### <font color = '#0f0264'> Introducerea filtrului capacitiv de tensiune </font>
Performanta redresorului poate fi imbunatatita prin introducerea intre bornele de iesire ale acestuia, in paralel cu sarcina $R_L$, a unui condensator de capacitate electrica mare ($C >> sute [\mu F]$), care are rolul de netezi variatiile de tensiune electrica. Schema electrica a redresorului monoalternanta cu filtru capacitiv este prezentata in Figura 3.

<img width = "300" height = "200" src="503.png" imgborder = "5">
<p style="text-align: center;">Figura 3. Redresorul monoalternanta cu filtru capacitiv</p>

Prin introducerea condensatorului de filtrare $C$ la iesirea redresorului, variatia tensiunii de iesire se reduce direct proportional cu valoarea capacitatii electrice $C$), asa cum se observa si in Figura 4. Pentru comparaţie, cu linie mai deschisă la culoare se prezintă şi forma de undă a tensiunii de ieşire a redresorului fără filtru. 

<img width = "500" height = "400" src="504.png" imgborder = "5">
<p style="text-align: center;">Figura 4. Formele de unda ale tensiunilor electrice ale redresorului monoalternanta cu filtru capacitiv. </p>

Se poate demonstra că, in ipoteza in care este satisfacuta conditia 

$$ R_L \cdot C >> T$$

unde T este perioada tensiunii de intrare $v_I$, atunci variaţia $\Delta v_O$ a tensiunii de ieşire a redresorului cu filtru capacitiv, numita si **riplul de tensiune**, este aproximativ:

$$ \Delta v_O \approx \frac{V_o} {f \cdot R_L \cdot C} $$  

unde:
- $V_o$ reprezinta valoarea maxima a tensiunii de iesire $v_O$,
- $f=\frac{1}{T}$ reprezinta frecventa tensiunii $v_I$; pentru redresoarele utilizate in sursele de alimentare din sistemele electronice UE, frecventa $f = 50 [Hz]$, iar $T = 50 [ms] = 0.05[s]$. 

Pentru redresorul monoalternanta cu filtru capacitiv, factorul de ondulatie depinde de valoarea capacitatii electrice a condensatorului de filtrare si se poate calcula cu relatia:

$$\gamma_{monoalternanta_C} = \frac{1}{2\cdot \sqrt{3} \cdot f \cdot R_L \cdot C}$$

## <font color = '#0f0264'> 1.2. Redresorul monofazat bialternanţă </font>
Redresorul monofozata bialternanta are 2 variante:
- in punte
- cu priza mediana

Deoarece prima varianta este mai utilizata, in Figura 5 se prezinta schema electrica a redresorului bialternanţă in punte. In aceasta varianta de redresor, functia de redresare a curentului electric este realizata de puntea de diode $D_1 - D_2, D_3 - D_4$, care asigura trecerea curentului electric prin sarcina $R_L$ in ambele alternante ale tensiunii de intrare $v_I$, spre deosebire de varianta monoalternanta, in care curentul electric este blocat in alternanta negativa a lui $v_I$.

<img width = "400" height = "300" src="505.png" imgborder = "5">
<p style="text-align: center;">Figura 5. Redresorul bialternanta</p>

La intrarea redresorului se aplică o tensiune sinusiodală $v_I(t)$ de amplitudine mare (în general de ordinul zecilor de volţi), a cărei componentă medie este nulă (variaţia tensiunii de intrare este axată pe zero volti). 

Cele 4 diode ale puntii redresoare sunt astfel conectate, incat functioneaza in perechi:
- pe durata semialternanţei pozitive a tensiunii de intrare (durata în care $v_I$>0[V]), diodele $D_1$ si $D_2$ funcţionează în conducţie directă, iar diodele $D_3$ si $D_4$ funcţionează în conducţie inversa; diodele $D_1$ si $D_2$
permit trecerea curentului electric, care ajunge pe rezistenta de sarcina $R_L$ şi generează pe aceasta o tensiune electrica a carei valoare tinde la valoarea tensiunii de intrare $v_I$; diferenta dintre valoarea tensiunii de intrare $v_I$ si valoarea tensiunii de iesire $v_O$ este data de tensiunea care cade pe rezistenta electrica a diodei, pe care se pierde o parte din valoarea tensiunii de intrare $v_I$;  
- pe durata semialternanţei negative a tensiunii de intrare (durata în care $v_I$<0[V]), diodele $D_1$ si $D_2$ funcţionează în conducţie inversa, iar diodele $D_3$ si $D_4$ funcţionează în conducţie directa; diodele $D_3$ si $D_4$
permit trecerea curentului electric, care ajunge pe rezistenta de sarcina $R_L$ şi generează pe aceasta o tensiune electrica a carei valoare tinde la valoarea tensiunii de intrare $v_I$ cu semn schimbat; diferenta dintre valoarea tensiunii de intrare $v_I$ si valoarea tensiunii de iesire $v_O$ este data de tensiunea care cade pe rezistenta electrica a diodei, pe care se pierde o parte din valoarea tensiunii de intrare $v_I$.

In final, se constata ca, in ambele alternante ale tensiunii de intrare $v_I$, se asigura un curent electric prin sarcina $R_L$, iar acest curent nu isi inverseaza sensul, ci si-l mentine in aceeasi directie.

Forma de undă a tensiunii electrice de iesire, obtinute prin redresare, este indicata in Figura 5. Pentru redresorul bialternanta, factorul de ondulatie se reduce la valoarea:

$$\gamma_{bialternanta} \approx 0.48$$ 

ceea ce indica o imbunatatire substantiala a performantei; in aceasta varianta de redresor, componenta se curent aletrnativ in teniunea de iesire este redusa considerabil, in comparatie cu varianta monoalternanta.

Se poate demonstra că, in ipoteza in care este satisfacuta conditia 

$$ R_L \cdot C >> T$$

unde T este perioada tensiunii de intrare $v_I$, atunci variaţia $\Delta v_O$ a tensiunii de ieşire a redresorului cu filtru capacitiv, numita si **riplul de tensiune**, este aproximativ:

# <font color = '#0f0264'>  2. Analiza in regim tranzitoriu a circuitelor electronice </font>

**Analiza in regim tranzitoriu permite determinarea variatiei valorii unei marimi electrice in timp**. 

In PySpice, analiza in regim tranzitoriu este realizata prin utilizarea liniei de cod indicata mai jos:

```python
    rezultate_analiza = instanta_simulator.transient(end_time = ...., step_time = ....)
```

unde:
- *end_time*: reprezinta un parametru al analizei, a carei valoare stabileste timpul final de simulare; in cazul in care marimile electrice din circuit variaza periodic, valoarea parametrului se calculeaza cu relatia urmatoare:

$$endtime = N \cdot PERIOD$$  

unde *PERIOD* reprezinta valoarea perioadei marimilor electrice din circuit, iar *N* reprezinta numarul de perioade pentru care se doreste efectuarea analizei; valoarea perioadei unei marimi electrice periodice se poate extrage prin utilizarea atributului `period`, aplicat asupra sursei de semnal electric periodic, care determina variatia marimilor electrice din circuit (in sectiunea 3 se explica modul de utilizare al acestui atribut). 

- *step_time*: reprezinta un parametru al analizei, a carei valoare stabileste pasul de simulare;valoarea parametrului se calculeaza cu relatia urmatoare:

$$steptime = \frac{endtime}{POINTS}$$  

unde *POINTS* reprezinta numarul de puncte de simulare;
- *rezultate_analiza*: reprezinta o variabila in care se salveaza rezultatele generate dupa efectuarea analizei.


**Observatie**: unitatea de masura a parametrilor analizei in regim tranzitoriu este secunda, care, in PySpice se exprima cu sintaxa `@u_s`.

Exemplu:
```python   
    # declatii constante:
    N      = 4        # numar perioade analiza
    PERIOD = 50@u_ms  # 50 milisecunde
    POINTS = 100      # numar puncte simulare
    
    # instanta simulator
    simulator  = circuit.simulator(temperature = 25, nominal_temperature= 25)
    
    # analiza in regim tranzitoriu
    # timp final simulare = N x PERIOD = 400 milisecunde
    # pas simulare = end_time / POINTS = 4 milisecunde
    analysis   = simulator.transient(end_time = N*PERIOD, step_time = N*PERIOD/POINTS )
```

# <font color = '#0f0264'> 3. Surse electrice sinusoidale </font>

Sursele de energie electrica pot genera intre cele 2 borne, `+`, respectiv `-`, o marime electrica a carei valoare variaza in timp, dupa o functie. In cazul in care aceasta este functia sinusoidala, se spune ca sursa respectiva este sinusoidala. In functie de marimea electrica generata, sursele de energie electrica sinusoidale sunt:
- sursa de tensiune sinusoidala
- sursa de curent sinusoidal

Sintaxa PySpice pentru generarea celor 2 tipuri de surse este indicata mai jos:


```python
    # sursa de tensiune sinusoidala:
    instanta_circ.SinusoidalVoltageSource(label, nod+,  nod-, 
                                          offset = ..., 
                                          amplitude = ..., 
                                          frequency = ...)
```

```python
    # sursa de curent sinusoidal:
    instanta_circ.SinusoidalCurrentSource(label, nod+,  nod-, 
                                          offset = ..., 
                                          amplitude = ..., 
                                          frequency = ...)
```

unde:
- *offset*: reprezinta un parametru care indica valoarea medie, sau valoare de curent continuu a variatiei marimii electrice generate; daca marimea electrica generata este tensiunea electrica, atunci unitatea de masura utilizata este `V` (VOLT), iar daca marimea electrica generata este curentul electric, atunci unitatea de masura utilizata este `A` (AMPER).
- *amplitude*: reprezinta un parametru care indica amplitudinea variatiei marimii electrice generate; daca marimea electrica generata este tensiunea electrica, atunci unitatea de masura utilizata este `V` (VOLT), iar daca marimea electrica generata este curentul electric, atunci unitatea de masura utilizata este `A` (AMPER).
- *frequency*: reprezinta un parametru care indica frecventa variatiei marimii electrice generate; unitatea de masura utilizata este `Hz` (Hertz).

Exemple:

```python
    # variabila V primeste instanta unei surse de tensiune sinusoidala, 
    # conectata intre nodurile 1 si 0 (borna '+' la nodul 1)
    # care genereaza o tensiune electrica de:
    # valoare medie = 0 volti
    # amplitudine   = 1 volt
    # frecventa     = 1kHz (1 kilohertz)
    V = circuit.SinusoidalVoltageSource('', 1, 0, 
                                        offset = 0@u_V, 
                                        amplitude = 1@u_V, 
                                        frequency = 1@u_kHz)
```
*

```python
    # variabila Ii primeste instanta unei surse de curent sinusoidal, 
    # conectata intre nodurile 0 si 1 (borna '+' la masa circuitului)
    # care genereaza un curent electric de:
    # valoare medie = 1 miliamper
    # amplitudine   = 500 microamperi
    # frecventa     = 5MHz (1 megahertzi)
    Ii = circuit.SinusoidalCurrentSource('i', 0, 1, 
                                         offset = 1@u_mA, 
                                         amplitude = 500@u_uA, 
                                         frequency = 5@u_MHz)
```

**Observatie**: valoarea **perioadei** unui semnal electric periodic (de exemplu, tensiune sinusoidala), este returnata prin aplicarea atributului `period` obiectului care reprezinta sursa care genereaza semnalul respectiv:

```python
    sursa.period
```

Exemplu:
```python
    V.period
```

unde *V* este sursa de tensiune sinusoidala definita ca in exemplul de mai sus, returneaza valoarea.... Atributul `period` este util pentru setarea valorilor analizei in regim tranzitoriu `transient`, `end_time`, respectiv `step_time`.  

Exemplu calcul parametri analiza in regim tranzitoriu:

```python   
    # declatii constante:
    N      = 4        # numar perioade analiza
    POINTS = 100      # numar puncte simulare
    
    # extragere perioada semnal sinusoidal:
    PERIOD = V.period  
    
    # instanta simulator
    simulator  = circuit.simulator(temperature = 25, nominal_temperature= 25)
    
    # analiza in regim tranzitoriu
    analysis   = simulator.transient(end_time = N*PERIOD, step_time = N*PERIOD/POINTS )
```

# <font color = '#0f0264'> 4. Determinarea variatiei marimilor electrice </font>

Variatia $\Delta x$ a unei marimi electrice notate generic cu $x$, se determina ca diferenta intre valoarea maxima $X_{max}$ a marimii electrice si valoarea minima  $X_{min}$ a marimii electrice:

$$ \Delta x = X_{max} - X_{min}$$

In cazul in care o marime electrica $x$ este reprezentata prin intermediul unui obiect de tip `WaveForm` (care stocheaza valoarea numerica si unitatea de masura corespunzatoare acesteia), valoarea maxima, respectiv minima, se extrag prin apelarea functiilor `max()`, respectiv `min()` asupra obiectului `WaveForm`:

```python
    # valoare maxima:
    x.max()  # return. val. max. a marimii electrice "x"
```

```python
    # valoare minima:
    x.min()  # return. val. min. a marimii electrice "x"
```
unde *x* este un obiect `WaveForm`.

Rezulta ca variatia unei marimi electrice reprezentate prin intermediul unui obiect `WaveForm` se poate determina astfel:

```python
    # variatia marimii electrice:
    x.max() - x.min() # return. variatia marimii electrice "x"
```

Exemplul 1: 
- determinarea variatiei unei tensiuni electrice determinate intre 2 puncte, A si B
- se cunosc valorile potentialelor electrice in nodurile A si B
```python
    # VA este potentialul electric in nodul A 
    # VB este potentialul electric in nodul B
    # VA, VB = obiecte de tip WaveForm
    # delta_V_AB este variatia tensiunii masurate intre A si B
    delta_V_AB = (VA - VB).max() - (VA - VB).min() 
```

Exemplul 2: 
- determinarea variatiei unei tensiuni electrice determinate intre 2 puncte, A si B
- se cunoaste valoarea tensiunii electrice in nodurile A si B
```python
    # V_AB este tensiunea electrica determinata intre punctele A si B
    # V_AB este un obiect de tip WaveForm
    # delta_V_AB este variatia tensiunii V_AB
    delta_V_AB = V_AB.max() - V_AB.min() 
```
