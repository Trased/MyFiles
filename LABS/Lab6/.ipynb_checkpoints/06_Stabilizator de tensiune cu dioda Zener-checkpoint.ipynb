## <font color = '#0f0264'> 1.1. Dioda Zener </font>

Dioda Zener este utilizată în special pentru implementarea stabilizatoarelor de tensiune, circuite electronice care intră în componenţa surselor de alimentare. 

Caracteristica de funcţionare a diodei Zener este prezentată în Figura 1. Spre deosebire de dioda semiconductoare, dioda Zener nu se distruge dacă funcţionează în regiunea de străpungere. Mai mult, în aplicaţii, dioda Zener este utilizată chiar în regiunea de străpungere în care, după cum se observă din caracteristica sa de funcţionare, **dioda Zener menţine între terminalele sale o tensiune electrica aproximativ constantă şi egală cu tensiunea de străpungere, notată $V_Z$, în condiţiile în care curentul electric prin aceasta prezinta variaţii semnificative de valori**. Un astfel de comportament este util în circuitele de stabilizare a valorii tensiunii electrice. Diodele Zener sunt produse diferite variante, care au tensiuni de străpungere de diverse valori, de la câţiva volţi (aplicaţiile de mică putere) până la sute de volţi (aplicaţii de mare putere).

## <font color = '#0f0264'> 1.2. Stabilizatorul de tensiune cu dioda Zener </font>

Rolul stabilizatorului de tensiune electrica (*voltage regulator*) este de a menţine tensiunea sa de iesire, notata $v_O$, la o valoare care, în mod ideal, este constantă, în condiţiile în care tensiunea electrica $v_I$, aplicată la intrarea sa, sau curentul electric $i_O$ care trece prin sarcina circuitului (un rezistor conectat între bornele de ieşire ale circuitului, sau alt circuit, conectat la iesirea stabilizatorului), sau temperatura de lucru T, pot varia între anumite limite. 

Performanţa stabilizatorului de tensiune electrica este descrisă prin intermediul unui set de parametri, care reflecta efectul pe care variaţia valorilor mărimilor indicate mai sus, il are în variaţia tensiunii de ieşire. 

Se definesc următorii parametri de performanţă ai stabilizatorului de tensiune, care, pentru un **stabilizator de tensiune performant**, trebuie să aibă **valori cât mai mici (ideal zero)**:

- factorul de stabilizare la variaţia tensiunii de intrare, notat $F_o$:

$$ F_o = \left | \frac{\Delta v_O}{\Delta v_I} \right |, \ daca \ i_O = constant, T = constant$$

- rezistenţa de ieşire a circuitului, notata $R_o$:

$$ R_o = \left | \frac{\Delta v_O}{\Delta i_O} \right |, \ daca \ v_I = constant, T = constant$$

- coeficientul de stabilizare cu temperatura, notat $K_T$:

$$ K_T = \left | \frac{\Delta v_O}{\Delta T} \right |, \ daca \ i_O = constant, v_O = constant$$

Elementul principal, care asigură menținerea la o valoare constantă a tensiunii la ieşirea stabilizatorului este dioda Zener. **Pentru menţinerea tensiunii de ieşire a stabilizatorului la o valoare constantă, este necesar ca dioda Zener să funcţioneze în <u>regiunea de străpungere</u>**. Dacă această condiţie este satisfăcută, dioda Zener menţine tensiunea pe terminalele sale la o valoare constantă şi egală cu tensiunea sa de străpungere, notată $V_Z$, pentru un domeniu de valori pentru curentul electric care trece prin dioda Zener, [$I_{ZMIN}$ ÷ $I_{ZMAX}$].

În Figura 2 este prezentată schema electronică a celui mai simplu stabilizator de tensiune electrica, realizat cu diodă Zener. 

<img width = "300" height = "200" src="602.png" imgborder = "5">
<p style="text-align: center;">Figura 2. Stabilizator de tensiune cu dioda Zener. </p>

Componenta electronica notata cu $D_Z$ reprezinta dioda Zener, elementul central al stabilizatorului de tensiune. Pentru ca aceasta sa stabilizeze valoarea tensiunii de iesire a circuitului la o valoare constanta, este necesar ca fie polarizata invers. Din acest motiv,  **anodul diodei se conecteaza la masa circuitului**, iar catodul la un punct de potential electric superior. 

Functionarea stabilizatorului de tensiune este caracterizata de 2 regiuni de functionare, demarcate in functie de regiunea in care dioda Zener functioneaza:
- **regiunea de stabilizare**: dioda Zener lucreaza in regiunea de strapungere; in consecinta, aceasta mentine la o valoare constanta tensiunea electric la terminalele sale si totodata valoarea tensiunii de iesire, in conditiile in care tensiunea de intrare $v_I$, sau curentul de iesire $i_O$, sau temperatura de lucru $T$ variaza;
- **regiunea interzisa**: dioda Zener lucreaza in conductie inversa, fara a se strapunge; in consecinta, valoare tensiunii electrice la terminalele sale variaza in functie de curentul care trece prin ea si din acest motiv, tensiunea de iesire a circuitului variaza, in conditiile in care tensiunea de intrare $v_I$, sau curentul de iesire $i_O$, sau temperatura de lucru $T$ variaza.

Pentru functionarea corecta a circuitului, dioda Zener trebuie conectata in circuit astfel incat sa fie polarizata invers. 
Daca valoarea rezistentei electrice $R_1$ a rezistorului care asigura polarizarea diodei este corect aleasa, atunci, dioda Zener ajunge sa lucreze in **regiunea de strapungere**, pentru un anumit domeniu de valori ale tensiunii de intrare $v_I$ si ale curentului electric $i_O$, solicitat de rezistorul de sarcina $R_L$. 

Se poate demonstra că, pentru menţinerea diodei Zener în regiunea de străpungere, valoarea rezistenţei $R_1$ trebuie să fie aleasă în domeniul de valori:

$$\left [ \frac{v_{IMAX} - V_Z}{I_{ZMAX} - i_{OMIN} }, \frac{v_{IMIN} - V_Z}{I_{ZMIN} - i_{OMAX} }  \right ]$$

unde: 
- $V_Z$, $I_{ZMIN}$ şi $I_{ZMAX}$ sunt parametrii de catalog ai diodei Zener considerate, 
- $v_{IMIN}$ ÷ $v_{IMAX}$, respectiv $i_{OMIN}$ ÷ $i_{OMAX}$ reprezintă domeniul de valori în care mărimile electrice $v_I$ şi $i_O$ pot varia, valori care sunt stabilite la proiectarea circuitului. 

## <font color = '#0f0264'> 1.3. Sursa de tensiune continua </font>

Sursa de tensiune continuă este utilizată pentru alimentarea circuitelor electronice cu o tensiune de valoare constantă. În mod ideal, valoarea tensiunii de alimentare trebuie să se menţină la o valoare constantă, indiferent de valoarea curentului solicitat sursei de alimentare de către circuitul alimentat, sau de variaţiile temperaturii la care lucrează circuitul respectiv. Sursa de tensiune cuprinde blocurile distincte prezentate în Figura 3. 


<img width = "800" height = "600" src="603.png" imgborder = "5">
<p style="text-align: center;">Figura 3. Sursa de tensiune continua. </p>

- **transformatorul**: rolul acestuia este de a reduce variaţia tensiunii sinusoidale furnizată de reteaua de distributie a energiei electrice la priza de alimentare, la o valoare adaptata la parametrii diodelor utilizate in redresorul de tensiune;
- **redresorul de tensiune**: rolul acestuia este de a converti tensiunea alternativă, aplicată la intrarea sa, într-o tensiune continuă pulsatorie; este realizat pe baza diodelor redresoare;
- **filtrul de tensiune**: are rolul de a reduce variaţiile tensiunii redresate; este realizat prin intermediul unui condensator de capacitate electrica mare (minim sute microfarazi);
- **stabilizatorul de tensiune**: are rolul de a menţine la bornele sale de ieşire o valoare constantă a tensiunii electrice, indiferent de variaţiile tensiunii de intrare, de valoarea curentului solicitat de circuitul alimentat sau de variaţiile temperaturii de lucru; cel mai simplu stabilizator de tensiune este realizat cu ajutorul unei diode Zener, care funcţionează în regiunea de străpungere;
- **sarcina**: reprezintă circuitul alimentat; acesta poate fi reprezentat prin intermediul unei rezistenţe echivalente, masurate între bornele de intrare ale circuitului respectiv. 

# <font color = '#0f0264'> 2. Analiza de curent continuu (*Direct Current Analysis*) in PySpice</font>

La executia acestei analize sunt calculate **variatiile** marimilor electrice ale circuitului, **tensiuni electrice**, respectiv ale **curenti electrici**, **in functie de alte marimi electrice** ale circuitului, la diverse temperaturi de lucru, in anumite intervale de variatie;


Aceasta analiza este utila pentru determinarea caracteristicilor de functionare ale componentelor electronice, de tip curent = functie(tensiune), sau a caracteristicilor de functionare ale circuitelor electronice, de tip marime electrica de iesire = functie(marime electrica de intrare).

In cadrul acestui laborator, analiza de curent continuu va fi realizata pentru a determina caracteristica de functionare a diodei Zener, $i_Z = f(v_Z)$, respectiv variatia tensiunii de iesire $v_O$ la variatia tensiunii de intrare $v_I$, respectiv variatia tensiunii de iesire $v_O$ la variatia curentului de iesire $i_O$:

1. $$i_Z = f(v_Z)$$
<br>
2. $$v_O = f(v_I)$$

Sintaxa PySpice generala, utilizata pentru setarea unei analize, este urmatoarea:

```python
    instanta_analiza = instanta_simulator.dc(nume_sursa = slice(start, end, step) )
```

unde:
- *instanta_analiza* reprezinta o variabila care stocheaza rezultatele si parametrii analizei efectuate;
- *dc* reprezinta analiza de curent continuu (DC = D*irect* C*urrent*)
- *nume_sursa* reprezinta numele sursei de tensiune continua, sau de curent continuu, a carei marime electrica generata este variata in cadrul analizei
- `slice(start, end, step)` reprezinta functia [`slice()`](https://www.w3schools.com/python/ref_func_slice.asp) implementata in Python, care genereaza pentru sursa de tensiune sau curent variata un domeniu de valori [*start*,*end*], cu un pas egal cu *step*.


Exemplul 1:

```python
    # valoarea tensiunii electrice generata de sursa de tensiune VI 
    # este variata in intervalul [0 - 10] volti, cu un pas de 1 volt
    
    # instanta sursa tensiune VI
    VI = circuit.V('I',  1, circuit.gnd,  0@u_V )
    ...
    # instanta simulator
    simulator  = circuit.simulator(temperature = 25, nominal_temperature= 25)
    
    # analiza
    analysis_results = simulator.dc(VI = slice(0, 10, 1))
```

Exemplul 2:
```python
    # valoarea curentului electric generat de sursa de curent Ii 
    # este variata in intervalul [0 - 1] amperi, cu un pas de 1 mA (miliamper)
    
    # instanta sursa curent Ii
    Ii = circuit.I('i', circuit.gnd,  1,  0@u_A )
    ...
    # instanta simulator
    simulator = circuit.simulator(temperature = 25, nominal_temperature= 25)
    
    # analiza
    analysis_results = simulator.dc(Ii = slice(0, 1, 0.001))
```
