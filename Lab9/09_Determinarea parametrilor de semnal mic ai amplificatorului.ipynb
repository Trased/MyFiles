## <font color = '#0f0264'> 1.1. Parametrii de semnal mic ai amplificatorului </font>

Figura 1 prezinta schema bloc a amplificatorului. Rolul acestuia este de a creşte puterea electrică a semnalului electric aplicat la bornele sale de intrare, fara a introduce distorsiuni in marimea furnizata la bornele sale de iesire.

<img width = "400" height = "300" src="901.png" imgborder = "5">
<p style="text-align: center;">Figura 1. Schema bloc a unui amplificator. </p>

În cazul în care amplificatorul este construit pe baza unui tranzistor bipolar, pentru evitarea apariţiei distorsiunilor în forma de undă a semnalului de ieşire, este necesar ca funcţionarea tranzistorului sa indeplineasca urmatoarele conditii:
- tranzistorul trebuie sa lucreze permanent în Regiunea Activă Normală (RAN). Pentru menţinerea funcţionării tranzistorului bipolar în RAN, trebuie satisfăcută condiţia:

$$0.5[V] < V_{CE} < V_{CC} - 0.5[V]$$

unde $V_{CE}$ este tensiunea continua intre colectorul si emitorul tranzistorului, iar $V_{CC}$ reprezinta valoarea tensiunii generate de catre sursa de alimentare cu energie electrica a circuitului;
- variatia tensiunii electrice intre baza si emitorul tranzistorului trebuie sa fie mai mica decat tensiunea termica $V_T$; in acest caz, se spune ca tranzistorul lucreaza in **regim de semnal mic**, caz in care comportamentul tranzistorului este considerat ca fiind aproximativ liniar.

In regim variabil de semnal mic, comportamentul amplificatorului in banda sa de frecventa este caracterizat de valorile parametrilor săi de semnal mic, valabili in banda de frecventa:
- factorul de amplificare, notat $A$,
- rezistența de intrare, notata $R_i$,
- rezistența de ieșire, notata $R_o$.

În cazul în care semnalele de intrare şi de ieşire ale amplificatorului sunt reprezentate prin intermediul tensiunilor electrice, factorul de amplificare de interes este factorul de amplificare in tensiune, notat $A_V$, care se determina cu relatia:

$$A_V = \frac{v_O}{v_I}$$

unde prin $v_O$ si $v_I$ s-au notat tensiunea de iesire a amplificatorului, respectiv tensiuniea sa de intrare. Modulul acestui parametru, notat $|A_V|$, furnizează informaţii despre nivelul creşterii amplitudinii semnalului aplicat la bornele de intrare şi definit prin relaţia generală:

$$|A_V| = \frac{V_o}{V_i}$$

unde prin $V_o$ si $V_i$ s-au notat amplitudinea tensiunii de iesire a amplificatorului $v_O$, respectiv a tensiunii de intrare $v_I$.

## <font color = '#0f0264'> 1.2. Conectarea amplificatorului la circuitele externe </font>

Valoarea amplificarii in tensiune, determinata dupa relatia indicata sus, este valabila numai daca amplificatorul **nu este conectat la circuite externe**. 

In aplicatiile practice, amplificatorul este conectat intotdeauna la o sursa de informatii, de la care amplificatorul primeste semnalul electric pe care il amplifica, respectiv la o sarcina, careia amplificatorul ii furnizeaza semnalul electric amplificat. 

In Figura 2 se prezinta modul in care este conectat amplificatorul la un generator de semnal (sursa de infotmatii), care genereaza o tensiune electrica $v_G$, respectiv la o sarcina $R_L$. Generatorul de semnal furnizeaza amplificatorului semnalul electric de amplificat, reprezentat de tensiunea electrica $v_G$, iar sarcina primeste semnalul amplificat de catre amplificator, tensiunea electrica $v_L$. In Figura 2, prin $R_g$ s-a notat rezistenta electrica echivalenta a generatorului de semnal $v_G$, masurata intre bornele de iesire ale acestuia.

<img width = "800" height = "600" src="902.png" imgborder = "5">
<p style="text-align: center;">Figura 2. Conectarea amplificatorului la generatorul de semnal si la sarcina. </p>

In acest caz, amplificarea in tensiune a intregului circuit, denumita si amplificarea reala in tensiune, notata $A_{Vg}$, devine:

$$A_{Vg} = \frac{v_L}{v_G}$$

iar modulul amplificarii reale in tensiune, se calculeaza cu relatia:

$$|A_{Vg}| = \frac{V_l}{V_g}$$

unde prin $V_l$ si $V_g$ s-au notat amplitudinea tensiunii $v_L$, masurata pe sarcina $R_L$ a amplificatorului, respectiv a tensiunii generate de catre generatorul de semnal, $v_G$.

### <font color = '#0f0264'> 1.2.1. Pierderile de semnal la intrarea amplificatorului</font>

Din cauza ca amplificatorul prezinta intre bornele sale de intrare o rezistenta de intrare $R_i$ finita, o parte din informatia reprezentata de tensiunea electrica $v_G$ se pierde pe rezistenta interna $R_g$ a generatorului de semnal. In consecinta, tensiunea care se aplica pe bornele de intrare ale amplificatorului este mai mica decat tensiunea electrica generata de catre generatorul de semnal:

$$v_I < v_G$$

Pierderea de informatie pe rezistenta interna $R_g$ a generatorului de semnal se poate reduce daca amplificatorul este astfel proiectat incat rezistenta sa de intrare sa rezulte la o valoare foarte mare:

$$R_i \rightarrow \infty [\Omega]$$

### <font color = '#0f0264'> 1.2.2. Pierderile de semnal la iesirea amplificatorului</font>

Din cauza ca amplificatorul prezinta intre bornele sale de iesire o rezistenta de iesire $R_o$ nenula, o parte din informatia amplificata de catre circuit, reprezentata de tensiunea electrica $v_O$, se pierde pe rezistenta de iesire $R_o$ a amplificatorului. In consecinta, tensiunea electrica furnizata sarcinii $R_L$ a amplificatorului este mai mica decat tensiunea electrica amplificata de catre amplificator si furnizata la bornele sale de iesire:

$$v_L < v_O$$

Pierderea de informatie pe rezistenta de iesire $R_o$ a amplificatorului se poate reduce daca amplificatorul este astfel proiectat, incat rezistenta sa de iesire sa rezulte la o valoare foarte mica: 

$$R_o \rightarrow 0[\Omega]$$

Din cauza pierderilor de semnal prezente la bornele de intrarea si de iesire ale amplificatorului, valoarea modulului amplificarii in tensiune a amplificatorului conectat la sursa de semnal $v_G$ si la sarcina $R_L$, $|A_{Vg}|$ este mai mica decat valoarea modulului amplificarii in tensiune a amplificatorului neconectat $|A_{V}|$:

$$|A_{Vg}| < |A_{V}|$$

Daca amplificatorul de tensiune este astfel proiectat incat 

$$R_i \rightarrow \infty [\Omega]$$
$$R_o \rightarrow 0[\Omega]$$

atunci pierderile de semnal la bornele de intrarea si de iesire ale amplificatorului se reduc si modulul amplificarii reale in tensiune tinde sa egaleze valoarea modulului amplificarii in tensiune a amplificatorului neconectat:

$$|A_{Vg}| \rightarrow |A_{V}|$$

## <font color = '#0f0264'> 1.3. Exemplu de amplificator cu tranzistor bipolar </font>


<img width = "500" height = "400" src="903.png" imgborder = "5">
<p style="text-align: center;">Figura 3. Amplificator de semnal mic cu tranzistor bipolar. </p>

Amplificatorul este realizat pe baza tranzistorului bipolar $Q$, care este polarizat de la sursa de tensiune $V_{CC}$ si rezistoarele $R_{B1}$, $R_{B2}$, $R_C$, $R_E$, astfel incat sa functioneze in Regiunea Activa Normala - $RAN$. 

Amplificatorul indicat in Figura 3 se numeste amplificator cu tranzistor in conexiunea Emitor Comun. In cazul in care amplificatorul lucreaza in regim de semnal mic ($V_{be} < 12.5[mV]$), comportamentul amplificatorului in banda sa de frecventa este descris de un set de parametri, numiti parametrii de semnal mic ai amplificatorului. Parametrii de semnal mic ai amplificatorului furnizeaza informatii despre modul in care circuitul amplifica, respectiv despre comportamentul amplificatorului la bornele de intrare, repsectiv de iesire. Parametrii de semnal mic ai amplificatorului, valabili in banda sa de frecventa, sunt determinati cu relatiile indicate mai jos:

- modulul amplificarii in tensiune in banda de frecventa (parametrul este notat si $A_{V0}$):

$$ |A_V| = \frac{\beta \cdot R_C}{r_\pi+(1+\beta)\cdot R_E} $$ 

- rezistenta de intrare:

$$ R_i = \frac{R_B \cdot (r_\pi + (1+\beta) \cdot R_E)}{R_B + r_\pi + (1+\beta) \cdot R_E} $$ 

unde:

$$ R_B = \frac{ R_{B1} \cdot R_{B2} }{R_{B1} + R_{B2}} $$ 

$$ r_{\pi} = \frac{ \beta \cdot V_T }{I_C} $$ 

unde:
 - $I_C$ este curentul continuu din colector (curentul determinat in Punctul Static de Functionare)
 - $\beta$ este factorul de amplificare in curent al tranzistorului
 - $V_T$ este tensiunea termica; in calcule, se considera valoarea tensiunii termice la temperatura camerei: $V_T = 25[mV]$ 


- rezistenta de iesire:

$$ R_o = R_C $$ 

- modulul amplificarii in curent in banda de frecventa (parametrul este notat si $A_{I0}$):

$$ |A_I| = |A_V| \cdot\frac{R_i}{R_o} $$ 

Dupa determinarea parametrilor de semnal mic ai amplificatorului, se pot determina valorile reale ale factorilor de amplificare ai acestuia, la conectarea cu circuitele externe. Formulele de calcul pentru factorii de amplificare reali (cei care sunt valabili la conectarea amplificatorului la circuitele externe), sunt determinate pe baza circuitelor echivalente, descrise de parametrii de semnal mic determinati.

Modulul amplificarii reale in tensiune a amplificatorului conectat la circuitele externe, se  determina cu relatia:

$$ |A_{Vg}| = \frac{R_i}{R_i + R_g} \cdot \frac{R_L}{R_L + R_o} \cdot |A_V| $$

Modulul amplificarii reale in curent a amplificatorului conectat la circuitele externe, se  determina cu relatia:

$$ |A_{Ig}| = \frac{R_g}{R_i + R_g} \cdot \frac{R_o}{R_L + R_o} \cdot |A_I| $$

Modulul amplificarii reale in curent a amplificatorului conectat la circuitele externe, in functie de modulul amplificarii reale in tensiune a amplificatorului conectat la circuitele externe, se determina cu relatia:

$$ |A_{Ig}| = |A_{Vg}| \cdot \frac{R_g}{R_L}$$

# <font color = '#0f0264'> 2. Determinarea amplitudinii variatiei marimilor electrice </font>

Amplitudinea variatiei $\Delta x$ a unei marimi electrice notate generic cu $x$, se determina ca diferenta intre valoarea maxima $X_{max}$ a marimii electrice si valoarea minima $X_{min}$ a marimii electrice, impartita la 2:

$$ X = (X_{max} - X_{min})/2$$

In cazul in care o marime electrica $x$ este reprezentata prin intermediul unui obiect de tip `WaveForm` (care stocheaza valoarea numerica si unitatea de masura corespunzatoare acesteia), valoarea maxima, respectiv minima, se extrag prin apelarea functiilor `max()`, respectiv `min()` asupra obiectului `WaveForm`. Rezulta ca amplitudinea variatiei unei marimi electrice reprezentate prin intermediul unui obiect `WaveForm` se poate determina astfel:

```python
    # amplitudinea variatiei marimii electrice:
    X = ( x.max() - x.min() ) / 2 # amplitudinea variatiei marimii electrice "x"
```

Exemplul 1: 
- determinarea amplitudinii variatiei unei tensiuni electrice determinate intre 2 puncte, A si B
- se cunosc valorile potentialelor electrice in nodurile A si B
```python
    # VA este potentialul electric in nodul A 
    # VB este potentialul electric in nodul B
    # VA, VB = obiecte de tip WaveForm
    # V_ab = amplitudinea variatiei tensiunii electrice, masurate intre A si B
    V_ab = ( (VA - VB).max() - (VA - VB).min() )  / 2 
```

Exemplul 2: 
- determinarea variatiei unei tensiuni electrice determinate intre 2 puncte, A si B
- se cunoaste valoarea tensiunii electrice in nodurile A si B
```python
    # v_AB este tensiunea electrica determinata intre punctele A si B
    # v_AB este un obiect de tip WaveForm
    # V_ab = amplitudinea variatiei tensiunii electrice, masurate intre A si B
    V_ab = ( v_AB.max() - v_AB.min() )  / 2 
```
