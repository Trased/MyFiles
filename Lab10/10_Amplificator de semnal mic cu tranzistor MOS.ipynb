## <font color = '#0f0264'> 1.1. Tranzistorul MOS </font>

Tranzistorul MOS (Metal Oxide Semiconductor) este un dispozitiv semiconductor, realizat din materiale semiconductoare de tip N si P, a carui structura este prezentata in Figura 1 (sunt posibile 2 versiuni constructive). 

<img width = "500" height = "400" src="1000.png" imgborder = "5">
<p style="text-align: center;">Figura 1. Structura tranzistorului MOS</p>

 Tranzistorul are 4 terminale, atasate la stuctura interna prin intermediul unor contacte metalice, denumite:
- DRENĂ (*Drain*), 
- GRILĂ (*Gate*)
- SURSĂ (*Source*)
- SUBSTRAT (*Bulk*)

Prin structura MOS, curentul este generat intre Drena si Sursa, daca sunt satisfacute anumite conditii de functionare, care conduc la aparitia in structura tranzistorului, intre materialele semiconductoare carora le sunt atasate cele 2 terminale mentionate, a unui "canal", care asigura suportul fizic pentru deplasarea de sarcina electrica intre cele 2 teminale. Canalul mentionat poate fi asigurat la fabricarea tranzistorului, caz in care tranzistorul MOS se numeste cu canal iniţial, sau in timpul utilizarii tranzistorului, caz in care tranzistorul MOS se numeste cu canal indus.

În funcţie de tipul materialului semiconductor din care este construit canalul, tranzistoarele MOS se împart în 2 tipuri şi anume:
-	tranzistoare MOS cu canal de tip N (canalul tranzistorului este realizat dintr-un material semiconductor de tip N, caz în care sarcinile electrice care se deplasează prin canal sunt constituite din electroni, particule care au sarcină electrică negativă);
-	tranzistoare MOS cu canal de tip P (canalul tranzistorului este realizat dintr-un material semiconductor de tip P, caz în care sarcinile electrice care se deplasează prin canal sunt constituite din goluri = particule fictive, care au sarcină electrică pozitivă);

În circuitele electronice, tranzistoarele MOS sunt simbolizate ca în Figura 2.

<img width = "500" height = "400" src="1001.png" imgborder = "5">
<p style="text-align: center;">Figura 2. Simbolurile electrice ale tranzistoarelor MOS</p>

La nivelul tranzistorului MOS apar 4 mărimi electrice: 
- 1 curent electric şi anume curentul care este generat între DRENĂ şi SURSĂ: 
  - $i_{D}$ – curentul de drenă 
- 3 tensiuni electrice şi anume tensiunile între terminalele tranzistoarelor:
  - $v_{GS}$ – tensiunea grilă-sursă 
  - $v_{DS}$ – tensiunea drenă-sursă 
  - $v_{GD}$ – tensiunea grilă-drenă


In functie de valorile tensiunilor $v_{GS}$ si $v_{DS}$, tranzistorul MOS poate functiona in mai multe regiuni de functionare. Singura regiune de functionare in care tranzistorul poate amplifica semnalele liniar (fara a le distorsiona), este **regiunea de saturatie**. 

Un tranzistor MOS lucreaza in regiunea de saturatie daca sunt satisfacute urmatoarele conditii de functionare:

$$v_{GS} \geq V_{TH}$$

$$v_{DS} \geq v_{GS} - V_{TH}$$

unde $V_{TH}$ reprezinta un parametru al tranzistorului MOS numit **tensiune de prag** si indica valoarea minima a tensiunii care trebuie aplicata intre Grila si Sursa tranzistorului, astfel incat prin acesta sa treaca un curent electric.  

## <font color = '#0f0264'> 1.2. Amplificator de semnal mic cu tranzistor MOS </font>

<img width = "500" height = "400" src="1002.png" imgborder = "5">
<p style="text-align: center;">Figura 3. Amplificator de semnal mic cu tranzistor MOS. </p>

Amplificatorul este realizat pe baza tranzistorului MOS $M$, cu canal indus de tip N, care este polarizat de la sursa de tensiune $V_{DD}$ si rezistoarele $R_{G1}$, $R_{G2}$, $R_D$,$R_S$, astfel incat sa functioneze in regiunea de saturatie. Condensatorul $C_S$ este utilizat pentru cresterea valorii amplificarii in tensiune.

Amplificatorul indicat in Figura 3 se numeste amplificator cu tranzistor MOS in conexiunea Sursa Comuna. Pe baza analizei comportamentului acestuia in regim variabil de semnal mic, se poate demonstra ca parametrii de semnal mic, care descriu comportamentul amplificatorului, se pot determina cu relatiile urmatoare:

- modulul amplificarii in tensiune:

$$ |A_V| = \frac{V_o}{V_i} $$ 

$$ |A_V| \approx g_m \cdot R_D $$ 

unde $V_o$ si $V_i$ reprezinta amplitudinea tensiunii $v_O$, respectiv amplitudinea tensiunii $v_I$.

- rezistenta de intrare:

$$ R_i = \frac{  R_{G1} \cdot   R_{G2}  }   { R_{G1} + R_{G2}  } $$ 
- rezistenta de iesire:

$$ R_o \approx R_D $$ 

Parametrul $g_m$ reprezinta panta tranzistorului si determina cu relatia:

$$ g_m = 2 \cdot \sqrt{k \cdot I_D} = 2 \cdot \sqrt{\frac{KP}{2} \cdot I_D}$$ 

unde $I_D$ este curentul continuu de drena (curentul din Punctul Static de Functionare al tranzistorului), iar $k$ reprezinta parametrul de transconductanta al tranzistorului MOS, iar $KP$ reprezinta parametrul de transconductanta al tranzistorului MOS, utilizat in modelul acestuia, care este descris in sectiunea 2.

Modulul amplificarii in curent, in banda de frecventa (parametrul este notat si $A_{I0}$), atunci cand amplificatorul nu este conectat la circuitele externe, se determina cu relatia:

$$ |A_I| = |A_V| \cdot\frac{R_i}{R_o} $$ 

## <font color = '#0f0264'> 1.3. Conectarea amplificatorului la circuitele externe</font>

<img width = "700" height = "900" src="1003.png" imgborder = "5">
<p style="text-align: center;">Figura 4.Conectarea amplificatorului la circuitele externe </p>

In Figura 4 se prezinta modul in care amplificatorul este conectat la circuitele externe:
- la bornele de intrare ale amplificatorului se conecteaza, prin intermediul unui condensator $C_G$, un generator de semnal, care genereaza o tensiune sinusoidala $v_G$; acesta furnizeaza intre bornele de intrare ale amplificatorului, tensiunea $v_I$ pe care acesta o amplifica; 
- la bornele de iesire ale amplificatorului se conecteaza, prin intermediul unui condensator $C_L$, o sarcina $R_L$; tensiunea amplificata de catre amplificator este livrata sarcinii sub forma tensiunii $v_L$; 

Dupa conectarea amplificatorului la circuitele externe, modulul amplificarii in tensiune se deterina cu relatia:

$$ |A_{Vg}| = \frac{V_l}{V_g} $$ 

unde $V_l$ si $V_g$ reprezinta amplitudinea tensiunii $v_L$, respectiv amplitudinea tensiunii $v_G$.

Din cauza rezistentei de intrare $R_i$ finite si a rezistentei $R_o$ nenule, la conectarea amplificatorului la circuitele externe, modulul amplificarii in tensiune scade. Modulul amplificarii reale in tensiune a amplificatorului se poate determina cu relatia urmatoare:

$$ |A_{Vg}| = \frac{R_i}{R_i + R_g} \cdot \frac{R_L}{R_L + R_o} \cdot |A_V| $$

Modulul amplificarii reale in curent a amplificatorului conectat la circuitele externe, se  determina cu relatia:

$$ |A_{Ig}| = \frac{R_g}{R_i + R_g} \cdot \frac{R_o}{R_L + R_o} \cdot |A_I| $$

sau, in functie de modulul amplificarii reale in tensiune a amplificatorului conectat la circuitele externe, se determina cu relatia:

$$ |A_{Ig}| = |A_{Vg}| \cdot \frac{R_g}{R_L}$$

## <font color = '#0f0264'> 1.4. Banda de frecventa a amplificatorului </font>

Valoarea modulului amplificării în tensiune $|A_{Vg}|$ este maxima in **bandă de frecvență** a amplificatorului; valoare amplificarii in tensiune care corespunde benzii de frecventa a amplificatorului se numeste **amplificare în bandă** și este notată $A_{Vg0}$. 

Banda de frecvență, notata $B$, este definita de diferenta dintre două frecvențe caracteristice ale amplificatorului, denumite **frecvență superioară**, notată $f_S$ , respectiv **frecvență inferioară**, notată $f_J$:

$$B = f_S - f_J$$

Frecvențele caracteristice ale amplificatorului sunt frecvențele de lucru ale acestuia la care valoarea modulului amplificării se reduce de $\frac{1}{\sqrt 2}$ ori față de nivelul amplificării în bandă $|A_{Vg0}|$. În figura de mai jos se indică graficul variației modulului amplificării în tensiune cu frecvența de lucru a amplificatorului, grafic denumit **caracteristica de frecvență a amplificatorului**, în care este scoasă în evidență banda de frecvență a amplificatorului.

# <font color = '#0f0264'> 2. Modelarea comportamentului tranzistorului MOS </font>

Modelul tranzistorului MOS contine o serie de parametri prin intermediul carora, comportamentul acestuia poate fi descris. Mai jos sunt indicati cativa parametri utilizati in modelarea tranzistorului MOS, modul in care sunt notati in PySpice precum si semnificatia acestora:

| Notatie PySpice| Unitate de masura| Semnificatie                                             |
|:--------------:|:----------------:|:--------------------------------------------------------:|
|VTO             | Volt             |tensiunea de prag a tranzistorului                        | 
|KP              | Amper / V^2      |parametru transconductanta tranzistor                     |
|RD              | Ohm              |rezistenta electrica din Drena tranzistorului             |
|RS              | Ohm              |rezistenta electrica din Sursa tranzistorului             |
|CBD             | Farad            |capacitatea electrica parazita substrat-drena             |
|CBS             | Farad            |capacitatea electrica parazita substrat-sursa             |

**Observatie**: in modelul tranzistorului MOS, parametrul de transconductanta $KP$ este egal cu dublul valorii parametrului de transconductanta al tranzistorului MOS, notat cu $k$, utilizat in formulele teoretice prezentate in curs:

$$K_P = 2\cdot k$$

## <font color = '#0f0264'> 2.1. Crearea modelului PySpice al tranzistorului NMOS (MOS cu canal N)</font>   

Sintaxa PySpice utilizata pentru crearea unui model al unui tranzistor MOS cu canal de tip N este urmatoarea:

```python
    instanta_circuit.model(nume_model, 'NMOS', level=1, parametrii_modelului_tranzistorului)
```

unde:
- *nume_model* reprezinta numele utilizat pentru identificarea modelului; acest nume va fi utilizat la crearea unei instante;
- *parametrii_modelului_tranzistorului* sunt parametrii modelului tranzistorului, ale caror valori individualizeaza tranzistorul modelat.   

Exemplu:

```python
    # model de tranzistor MOS 
    circuit.model('my_nmos', 'NMOS', level=1
              VTO = 1@u_V, 
              KP  = 400e-6, 
              CBD = 4@u_pF,
              CBS = 4@u_pF,
              RD  = 2@u_Ohm,
              RS  = 1.25@u_Ohm           
              )
```

## <font color = '#0f0264'> 2.2. Instantierea tranzistorului MOS in circuit </font> 

Sintaxa PySpice utilizata pentru instantierea unui tranzistor MOS in circuit este urmatoarea:

```python
    nume_tranzistor = instanta_circuit.MOSFET(label, nod_drena, nod_grila, nod_sursa, nod_substrat, model = nume_model)
```

unde *nod_drena*, *nod_grila*, *nod_sursa* si *nod_substrat* sunt nodurile de circuit in care este conectata Drena, Grila, Sursa si Substratul tranzistorului MOS, iar *nume_model* este numele modelului utilizat la crearea sa (in campul *nume_model* al definitiei modelului tranzistorului MOS).

**Observatie**: in circuitele studiate, Substratul va fi intotdeauna conectat la nodul la care este conectata Sursa tranzistorului MOS.

Exemplu:

```python
    # instanta tranzistor MOS
    # Drena este conectata in nodul 2
    # Grila este conectata in nodul 1
    # Sursa si Substratul sunt conectate la masa circuitului
    M = circuit.MOSFET('', 2, 1, 0, 0, model = 'my_nmos')
```
