## <font color = '#0f0264'> 1.1. Rolul circuitului de amplificare </font>

Schema de principiu a unui circuit de amplificare este indicata in Figura 1. Amplificatorul are rolul de a creşte puterea electrică a semnalului electric aplicat la bornele sale de intrare, de la o sursa de informatie externa - generatprul de semnal $v_G$. Amplificatorul creşte nivelul puterii electrice a semnalului electric pe baza energiei electrice preluate de la sursa sa de alimentare şi furnizează rezultatul obţinut pe o sarcina $R_L$, conectata la bornele sale de ieșire.

<img width = "750" height = "650" src="801.png" imgborder = "5">
<p style="text-align: center;">Figura 1. Schema de principiu a circuitului de amplificare. </p>

Amplificatorul trebuie să fie astfel proiectat incat să furnizeze la bornele sale de ieşire, conţinutul integral al informaţiei aplicate la intrarea sa. Pentru aceasta, semnalul generat la bornele de ieşire trebuie să aibă aceeaşi formă de undă cu cea a semnalului aplicat la bornele de intrare. Oricare modificare a formei de undă a semnalului la ieşire, în comparaţie cu cea a semnalului aplicat la intrare, determină distorsionarea (modificarea) conţinutului informaţiei prelucrate.  

În cazul în care amplificatorul este construit pe baza unui tranzistor bipolar, pentru evitarea apariţiei distorsiunilor în forma de undă a semnalului de ieşire, este necesar ca funcţionarea tranzistorului sa indeplineasca urmatoarele conditii:
- tranzistorul trebuie sa lucreze permanent în Regiunea Activă Normală (RAN). Pentru menţinerea funcţionării tranzistorului bipolar în RAN, trebuie satisfăcută condiţia:

$$0.5[V] < V_{CE} < V_{CC} - 0.5[V]$$

unde $V_{CE}$ este tensiunea continua intre colectorul si emitorul tranzistorului, iar $V_{CC}$ reprezinta valoarea tensiunii generate de catre sursa de alimentare cu energie electrica a circuitului;
- variatia tensiunii electrice intre baza si emitorul tranzistorului trebuie sa fie mai mica decat tensiunea termica $V_T$; in acest caz, se spune ca tranzistorul lucreaza in regim de semnal mic, caz in care comportamentul tranzistorului este considerat ca fiind aproximativ liniar.

## <font color = '#0f0264'> 1.2. Amplificarea circuitului </font>

În cazul în care informatia este reprezentata prin intermediul tensiunii electrice, parametrul care descrie functia de amplificare a circuitului de amplificare este reprezentat de catre **factorul de amplificare în tensiune**, notat $A_{Vg}$ şi definit prin relaţia generală:

$$A_{Vg} = \frac{v_L}{v_G}$$

Relatia intre amplitudinile celor 2 tensiuni $v_G$ si $v_L$, este furnizata de catre modulul amplificarii in tensiunea a circuitului, $|A_{Vg}|$:

$$|A_{Vg}| = \frac{V_l}{V_g}$$

unde $V_l$ si $V_g$ reprezinta amplitudinea tensiunii $v_L$, respectiv $v_G$.

## <font color = '#0f0264'> 1.3. Banda de frecventa a amplificatorului </font>

Valoarea modulului amplificării în tensiune $|A_{Vg}|$  depinde de frecvenţa semnalului aplicat la bornele de intrare ale amplificatorului, care stabilește totodată frecvența de lucru a acestuia. Există un domeniu de valori pentru frecvenţa de lucru, denumit **bandă de frecvență** a amplificatorului, în care valoarea $|A_{Vg}|$ este maxima, valoare amplificarii in tensiune care corespunde benzii de frecventa a amplificatorului se numeste **amplificare în bandă** și notată $A_{Vg0}$. În exteriorul benzii de frecvență, valoarea modulului amplificării în tensiune scade spre zero, în funcție de frecventa de lucru a amplificatorului. Din acest motiv se spune că amplificatorul amplifică numai în banda sa de frecvență.

Banda de frecvență, notata $B$, este definita de două frecvențe caracteristice ale amplificatorului, denumite **frecvență inferioară**, notată $f_J$, respectiv **frecvență superioară**, notată $f_S$:

$$B = f_S - f_J$$

Prin definiție, frecvențele caracteristice ale amplificatorului sunt frecvențele de lucru ale amplificatorului la care valoarea modulului amplificării se reduce de $\frac{1}{\sqrt 2}$ ori față de nivelul amplificării în bandă $|A_{Vg0}|$. În figura de mai jos se indică graficul variației modulului amplificării în tensiune cu frecvența de lucru a amplificatorului, grafic denumit **caracteristica de frecvență a amplificatorului**, în care este scoasă în evidență banda de frecvență a amplificatorului.

# <font color = '#0f0264'>  2. Analiza in regim variabil a circuitelor electronice </font>

**Analiza in regim variabil de semnal mic permite determinarea variatiei valorii unei marimi electrice in frecventa**. 

In PySpice, analiza in regim variabil este realizata prin utilizarea liniei de cod indicata mai jos:

```python
    rezultate_analiza = instanta_simulator.ac(start_frequency = ..., 
                                              stop_frequency = ...,
                                              variation = ...,
                                              number_of_points = 
                                             )
```

unde:
- *start_frequency* si *stop_frequency*: reprezinta 2 parametri ai analizei, care definesc domeniu de frecventa $[start-frequency, stop-frequency]$ in care se realizeaza analiza; variatia marimilor electrice ale circuitului se va realiza in domeniul astfel definit;

- *variation*: reprezinta un parametru al analizei, a carei valoare stabileste modul in care se va realiza analiza; pentru toate analizele de tipul `ac` realizate in acest laborator, acest parametru va primi valoarea `'dec'`, ceea ce indica o analiza pe decade (decada = 2 puteri consecutive ale lui 10);

- *number_of_points*: reprezinta un parametru care indica numarul de puncte de analiza considerat pe unitatea de variatie, precizata in parametrul *variation*; daca *variation =  'dec'*, atunci indica numarul de puncte de analiza pe decada.in care se salveaza rezultatele generate dupa efectuarea analizei.


**Observatie**: unitatea de masura a parametrilor analizei in regim variabil este Hertz-ul, care, in PySpice se exprima cu sintaxa `@u_Hz`.

Exemplu:
```python   
    # declatii constante:
    POINTS = 10       # numar puncte simulare/decada
    
    # instanta simulator
    simulator  = circuit.simulator(temperature = 25, nominal_temperature= 25)
    
    # analiza in regim variabil
    # interval analiza: 0.01Hz - 100MHz
    # analiza pe decade
    # numar puncte analiza/decada = POINTS
    
    analysis = simulator.ac(start_frequency = 0.01@u_Hz, 
                            stop_frequency=100@u_MHz,
                            variation = 'dec', 
                            number_of_points = POINTS
                           )

```

### <font color = '#0f0264'> Rezultatele generate prin analiza in regim variabil. </font>

In urma analizei in regim variabil, valorile marimilor electrice sunt returnate sub forma de numere complexe, care contin informatii atat despre magnitudinea marimii electrice cat si despre faza acesteia. 

Valoarea numerica a marimii electrice, calculate in urma analizei in regim variabil, se determina prin calcularea modulului numarului complex prin care este reprezentata marimea electrica:

$$|x| = \sqrt{{Re_x}^2 + {Im_x}^2}$$

Relatia de mai sus se poate exprima in Python prin apelarea functiei `sqrt()`, implementata in libraria `numpy`, respectiv aplicarea atributelor `real` si `imag` obiectului `WaveForm`, prin care se reprezinta marimea electrica de interes.

In exemplul de mai jos se indica modul in care se poate determina valoarea numerica a unei tensiuni electrice, calculate prin analiza in regim variabil:

```python
    # import librarie numpy
    import numpy as np

    # extragere WaveForm vx:
    vx = analysis.nodes['x']

    # calcul variatie marime electrica vx:
    delta_vx = np.sqrt(vx.real**2 + vx.imag**2 )
```
